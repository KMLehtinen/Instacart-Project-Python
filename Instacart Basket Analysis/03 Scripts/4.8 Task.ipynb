# 4. Loyalty Flag for existing customers

```python
# Creating a new column with 'max_order' grouped by 'user_id'
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Loyal Customer
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# Regular Customer
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# New Customer
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Checking head of ords_prods_merge
ords_prods_merge.head()
```

# 5. Spending Habits of the 3 types of customers

```python
# Check frequency of values in new loyalty flag column
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
#Aggregating prices based on loyalty_flag
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

#### The new customer actually spends most money on products. The loyal customers pend the least amount of money by about $3.00.

# 6. Targeting different types of spenders

```python
# Creating a new column with 'mean_product_price' grouped by 'user_id and prices'
ords_prods_merge['mean_product_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
# Low Spender
ords_prods_merge.loc[ords_prods_merge['mean_product_price'] < 10, 'spending flag'] = 'Low Spender'
```

```python
# High Spender
ords_prods_merge.loc[ords_prods_merge['mean_product_price'] >= 10, 'spending flag'] = 'High Spender'
```

```python
# Check frequency of values in new spending flag
ords_prods_merge['spending flag'].value_counts(dropna = False)
```

```python
ords_prods_merge.head()
```

# 7. Frequent vs. Non frequent Customers

```python
# Creating a new column: Customer type (non-frequent, regular, and frequent)
ords_prods_merge['customer_type'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Non-frequent Customer
ords_prods_merge.loc[ords_prods_merge['customer_type'] >20, 'Frequency flag'] = 'non-frequent customer'
```

```python
# Regular customer
ords_prods_merge.loc[(ords_prods_merge['customer_type'] > 10) & (ords_prods_merge['customer_type'] <= 20), 'Frequency flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['customer_type'] <10, 'Frequency flag'] = 'Frequent customer'
```

```python
# Check frequency of values in new frequency flag
ords_prods_merge['Frequency flag'].value_counts(dropna = False)
```

```python
ords_prods_merge.loc[ords_prods_merge['prices'] >100, 'prices'] = np.nan
```
